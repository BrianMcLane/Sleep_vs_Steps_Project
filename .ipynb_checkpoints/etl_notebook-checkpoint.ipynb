```python
import os
import pandas as pd
import numpy as np
import sqlalchemy
```

```python
daily_act_path = "fitbit_data/dailyActivity_merged.csv"
daily_act = pd.read_csv(daily_act_path)
daily_act
```

```python
daily_act.dtypes
```

```python
sleep_day_path = "fitbit_data/sleepDay_merged.csv"
sleep_day = pd.read_csv(sleep_day_path)
sleep_day[["date", "time", "am/pm"]] = sleep_day.SleepDay.str.split(" ", expand=True)
sleep_day
```

```python
sleep_day.dtypes
```

```python
fitbit = pd.merge(daily_act, sleep_day, left_on=['Id', 'ActivityDate'], right_on = ['Id', 'date'])
```

```python
fitbit.columns
```

```python
fitbit_clean = fitbit[['Id', 'ActivityDate', 'TotalSteps', 'TotalDistance',
       'LoggedActivitiesDistance', 'VeryActiveDistance',
       'ModeratelyActiveDistance', 'LightActiveDistance',
       'SedentaryActiveDistance', 'VeryActiveMinutes', 'FairlyActiveMinutes',
       'LightlyActiveMinutes', 'SedentaryMinutes', 'Calories', 'TotalMinutesAsleep', 'TotalTimeInBed']]
```
